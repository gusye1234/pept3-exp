```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
from collections import defaultdict
plt.rcParams['svg.fonttype'] = 'none'
```

```python
alleles_rawfile = {}
with open("./data/allele_raw.txt") as f:
    for l in f:
        pack = l.strip().split("\t")
        alleles_rawfile[pack[0]] = set(pack[1:])
```

```python
Alleles = sorted(alleles_rawfile.keys())
set_threshold = 0.1
frag_model = "prosit_l1"
threshold = 0.01

lsgs = []
for which in Alleles:
    ft_tab = f"/data1/yejb/prosit/figure3/percolator_hdf5_allele_{set_threshold}_{frag_model}/{which}"
    noft_tab = f"/data1/yejb/prosit/figure3/forPRIDE/Alleles/{which}/percolator"
    ft_p = pd.read_csv(os.path.join(ft_tab, "prosit_target.peptides"), sep='\t')
    noft_p = pd.read_csv(os.path.join(noft_tab, "prosit_target.peptides"), sep='\t')
    
    f = set(ft_p[ft_p['q-value'] < threshold]['peptide'].apply(
        lambda x: x.strip("_").strip(".")).unique())
    n = set(noft_p[noft_p['q-value'] < threshold]['peptide'].apply(
        lambda x: x.strip("_").strip(".")).unique())
    lsgs.append((which, (len(n-f), len(n.intersection(f)), len(f - n))))
    
    
```

```python
def plot_alleles_lsg(all_lsg):
    all_lsg = sorted(all_lsg, key=lambda x: (x[1][1] + x[1][2]))
    c_loss = 'orange'
    c_shared = 'darkslateblue'
    c_gain = 'green'
    linewidth = 0.6
    barwidth = 0.9
    fig, ax = plt.subplots(figsize=(9, 6), dpi=100)

    alleles = [x[0].split("_")[0] for x in all_lsg]
    norm_sum = np.array([x[1][0] + x[1][1] for x in all_lsg])
    losses = np.array([-x[1][0] for x in all_lsg])
    shares = np.array([x[1][1] for x in all_lsg]) 
    gains = np.array([x[1][2] for x in all_lsg]) 
    x_ticks = list(range(len(alleles)))
    p1 = ax.bar(alleles, losses, barwidth, label="Loss",
                color=c_loss, linewidth=linewidth, edgecolor='black')
    p2 = ax.bar(alleles, shares, barwidth, label="Shared",
                color=c_shared, linewidth=linewidth, edgecolor='black')
    p3 = ax.bar(alleles, gains, barwidth, label="Gain", bottom=shares,
                color=c_gain, linewidth=linewidth, edgecolor='black')
    # for i, allele in enumerate(alleles):
    #     ax.text(i-0.2, 40000, str(gains[i]), rotation=90, va='center')
    #     ax.text(i-0.2, 20000, str(shares[i]), rotation=90, va='center')
    #     ax.text(i - 0.2, -3000, str(abs(losses[i])), rotation=90, va='center')
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.set_ylabel("Number of identified Peptides", fontsize=15)
    # ax.spines['left'].set_visible(False)
    # ax.axes.get_xaxis().set_visible(False)
    return fig, ax
```

```python
fig, ax = plot_alleles_lsg(lsgs)
ax.set_title("Fine-tuned Prosit 2019 verus. Prosit 2020", fontsize=15)
ax.legend(loc='upper center',
          fancybox=True, shadow=True, ncol=5)
ax.set_ylim([-500, 7000])
ax.set_yticks([i * 1000 for i in range(8)])
ax.spines['left'].set_bounds(0, 7000)
ax.axhline(0, color='black', linewidth=0.8, linestyle='--', xmin=0, xmax=0.955)
ax.tick_params(axis='x', labelsize=7, rotation=60)
# ok = ax.set_xticklabels(ax.get_xticks(), rotation = 90)
# fig.savefig("./figs/alleles-venn-fp-p.svg", dpi=300, bbox_inches="tight")
```
