```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn3, venn3_circles
import pandas as pd

plt.rcParams['svg.fonttype'] = 'none'
```

```python
maxquant_peps = set(p.upper() for p in pd.read_csv("data/fig3_prosit_mel15_mut_maxquant.csv")['Modified Sequence'])
prosit_peps = set(p.upper() for p in pd.read_csv("data/fig3_prosit_mel15_mut.csv")['Sequence'])
ft_peps = set(p.upper() for p in pd.read_csv(
    "data/fig3_mut_ft_q_value.csv")['Sequence'])
```

```python
print(len(maxquant_peps), len(prosit_peps), len(ft_peps))
```

```python
plt.subplots(dpi=200)
# parts = (10, 8, 22, 6, 9, 4, 2)
parts = [maxquant_peps, prosit_peps, ft_peps]
v = venn3(subsets=parts,
          set_labels=('Maxquant', 'Prosit', 'FT Prosit'), alpha=0.7)
c = venn3_circles(subsets=parts,
                  linestyle='dashed', linewidth=1, color="k")
plt.title("Mutated peptide identifications \nfor Mel-15 with patient specific fasta", fontsize=15)
plt.savefig("figs/fig3_mut_venn.svg", bbox_inches="tight", dpi=300)
plt.show()
```
